```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
```

```python
df=pd.read_csv('test.csv',delimiter=',')
```

```python
X_test=df.iloc[:,0:-2]
```

```python
print((X_test.shape))
```

```python
Y_test=df.iloc[:,-1].values
print(Y_test.shape)
```

```python
df2=pd.read_csv('train.csv')
```

```python
X_train=df2.iloc[:,0:-2].values
Y_train=df2.iloc[:,-1].values
```

```python
print(X_train.shape)
```

```python
from sklearn.metrics import classification_report
```

```python
clf1 = RandomForestClassifier(n_estimators=1000)
clf1.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf1.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf1.score(X_test, Y_test)))
Y_test_pred = clf1.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

```python
clf2=KNeighborsClassifier(n_neighbors=12)
clf2.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf2.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf2.score(X_test, Y_test)))
Y_test_pred = clf2.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

```python
clf3=DecisionTreeClassifier()
clf3.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf3.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf3.score(X_test, Y_test)))
Y_test_pred = clf3.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

```python
clf4=LogisticRegression(max_iter=10000)
clf4.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf4.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf4.score(X_test, Y_test)))
Y_test_pred = clf4.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

```python
clf5=SVC(gamma='auto',kernel='rbf',C=1)
clf5.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf5.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf5.score(X_test, Y_test)))
Y_test_pred = clf5.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

```python
clf6=GaussianNB()
clf6.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf6.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf6.score(X_test, Y_test)))
Y_test_pred = clf6.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

```python
clf7=ExtraTreesClassifier()
clf7.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf7.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf7.score(X_test, Y_test)))
Y_test_pred = clf7.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

```python
clf8=BaggingClassifier()
clf8.fit(X_train, Y_train)
print("Accuracy on training set is : {}".format(clf8.score(X_train, Y_train)))
print("Accuracy on test set is : {}".format(clf8.score(X_test, Y_test)))
Y_test_pred = clf8.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```
